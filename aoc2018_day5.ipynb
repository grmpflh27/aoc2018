```go
func min(a, b int) int {
    if a < b {
        return a
    }
    return b
}

func max(a, b int) int {
    if a > b {
        return a
    }
    return b
}
```

```go
func react(polymer string) string {
    var THE_DIFF float64 = 32

    for{
        cur_len := len(polymer)

        var prev_char rune = -1
        for i, c := range polymer{

            if i == 0{
                prev_char = c
                continue
            }

            diff := math.Abs(float64(prev_char) - float64(c))

            if diff == THE_DIFF{
                // if found remove the 2 and start again from the beginning

                begin_idx := max(0, i-1)
                end_idx := min(i+1, cur_len - 1)

                polymer = fmt.Sprintf("%v%v", polymer[:begin_idx], polymer[end_idx:])
                break
            }

            prev_char = c
        }

        // do this until reaching the end
        if len(polymer) == cur_len{
            break
        }
    }
    
    return polymer
}


polymer = react(polymer)

fmt.Println("Length of reacted polymer", len(polymer) - 1)
```

```go
var polymer string = "dabAcCaCBAcCcaDA"
```

```go
func SnipChar(char int, cur string) string{
    capped := strings.Replace(cur, string(char), "", -1)
    capped = strings.Replace(capped, string(char + 32), "", -1)
    return capped
}


shortest := len(polymer)

for i:=65; i <= 65+26; i++{
    capped := SnipChar(i, polymer)
    reacted := react(capped)
    
    cur_len := len(reacted)
    
    if cur_len < shortest{
        shortest = cur_len
    }
}

fmt.Printf("%v is shortest\n", shortest - 1)
```
