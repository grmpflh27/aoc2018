```go
type ts struct { 
    month int
    day int
    hour int
    minute int
}


func ParseTimestamp(rec string) ts {
    var tsRegex = regexp.MustCompile(`^\[\d+-(\d+)-(\d+)\s{1}(\d+):(\d+)\].*$`)
    var match = tsRegex.FindStringSubmatch(rec)
    
    _mth, _ := strconv.Atoi(match[1])
    _dy, _ := strconv.Atoi(match[2])
    _hh, _ := strconv.Atoi(match[3])
    _mm, _ := strconv.Atoi(match[4])
    
    if _hh == 23{
        _mm = 0
        _hh = 0
    }
         
    cur_ts := ts{_mth, _dy, _hh, _mm}
    
    return cur_ts 
}
```

```go
func GetAllGuards(records []string) []string{
    
    var guards = map[string]int{}
    
    var guardRegex = regexp.MustCompile(`^.*Guard\s+#(\d+).*$`)
    for _, rec := range records{
        if strings.Contains(rec, "Guard"){
            var match = guardRegex.FindStringSubmatch(rec)[1]
            guards[match] = 1
        }
    }
    
    keys := []string{}
    for key, _ := range guards {
        keys = append(keys, key)
    }
    
    return keys
}

func initGuardSleepySchedules([]string) map[string][]float64{
    var guards = map[string][]float64{}
    for _, cur_guard := range guard_ids{
        mins := mat.NewVecDense(60, nil)
        guards[cur_guard] = mins.RawVector().Data
    }
    
    return guards
}

guard_ids := GetAllGuards(records)
guards := initGuardSleepySchedules(guard_ids)
```

```go
// 1) iterate over all records

// get span of this guard
exit := len(records)

var idx int = 0
var cur_guard_id string
var sleep_start, sleep_stop int = -1, -1

for {

    rec := records[idx]
    if strings.Contains(rec, "Guard"){
        var guardRegex = regexp.MustCompile(`^.*Guard\s+#(\d+).*$`)
        cur_guard_id = guardRegex.FindStringSubmatch(rec)[1]
        sleep_start = -1
        sleep_stop = -1
    } else {
        cur_ts := ParseTimestamp(rec)
        if strings.Contains(rec, "falls asleep"){
            sleep_start = cur_ts.minute
        } else if strings.Contains(rec, "wakes up"){
            sleep_stop = cur_ts.minute
        }
    }
    
    if sleep_start != -1 && sleep_stop != -1{
        for i := sleep_start; i < sleep_stop; i++{
            guards[cur_guard_id][i] += 1
        }
    }
    
    idx++
    
    if idx >= exit{
        break
    }
    
}
```

```go
func FindSleepiestGuard(map[string][]float64) string{
    var cur_max float64 = -1
    var sleepy_guard string
    for k, sleep_time := range guards{
        cur_sum := mat.Sum(mat.NewVecDense(len(sleep_time), sleep_time))
        if cur_sum > cur_max{
            cur_max = cur_sum
            sleepy_guard = k
        }
    }  
    
    return sleepy_guard
}

func GetBestMinute(sleep_schedule []float64) int{
    var best_min = -1
    var cur_max float64 = -1
    for i, cur := range sleep_schedule{
        if cur > cur_max{
            cur_max = cur
            best_min = i
        }
    }
    
    return best_min
}


```

```go
sleeper := FindSleepiestGuard(guards)
best_min := GetBestMinute(guards[sleeper])

//final result
guard_id_as_int, _ := strconv.Atoi(sleeper)
fmt.Printf("Guard #%v is the sleepiest. Best minute is %v. Final result %v\n", guard_id_as_int, best_min, guard_id_as_int * best_min)
```
