```go
import (
    "fmt"
    "regexp"
    "regexp/syntax"
    "strconv"
    "strings"
    "sort"
    "gonum.org/v1/gonum/mat"
    "math"
    "gonum.org/v1/gonum/floats"
)
```

```go
func ManhattanDistance(p, q []float64) int{
    erg := [] float64{}
    for i :=0; i < len(p); i++{
        erg = append(erg, math.Abs(p[i] - q[i]))
    }
    
    return int(mat.Sum(mat.NewVecDense(len(erg), erg)))
}
```

```go
coord_str := `1, 1
1, 6
8, 3
3, 4
5, 5
8, 9`
```

```go
coords := strings.Split(coord_str, "\n")
```

```go
coord_regex := regexp.MustCompile(`^(\d+),\s+(\d+)$`)


var mapping = map[float64] []float64{}

var min_x, min_y, max_x, max_y float64 = 1000, 1000, 0, 0

for id, coord := range coords{
    //parse
    cur := coord_regex.FindStringSubmatch(coord)[1:]
    coord_1, _ := strconv.ParseFloat(cur[0], 10)
    coord_2, _ := strconv.ParseFloat(cur[1], 10)
    
    // get span of field
    if coord_1 < min_x{
        min_x = coord_1
    } else if coord_1 > max_x{
        max_x = coord_1
    }
        
    if coord_2 < min_y{
        min_y = coord_2
    } else if coord_2 > max_y{
        max_y = coord_2
    }
    
    mapping[float64(id) + 1] = []float64{coord_1, coord_2}
}

fmt.Println(mapping)
fmt.Println("Extent", min_x, max_x, min_y, max_y)
```

```go
field := mat.NewDense(int(max_y) + 1, int(max_x) + 1, nil)
```

```go
// init field

for i, coord := range mapping{
    field.Set(int(coord[1]), int(coord[0]), i)
}
```

```go
func printField(field *mat.Dense){
    r, _ := field.Dims()
    for i:=0; i < r; i++{
        fmt.Printf("%v\n", field.RowView(i))
    }
}

func countMin(arr []float64) int{
    cnt := 0
    
    minArr := floats.Min(arr)
    for _, cur := range arr{
        if cur == minArr{
            cnt++
        }
    }
    return cnt
}
```

```go
func FindClosestPoint(x, y int) float64{
    smallest_distance := 10000
    closest_id := float64(-1)
    dists := []float64{}
    
    for cur_id, coord := range mapping{

        dist := ManhattanDistance(coord, []float64{float64(y), float64(x)})
        dists = append(dists, float64(dist))
        if dist < smallest_distance{
            smallest_distance = dist
            closest_id = cur_id
        }
    }
    //fmt.Printf("Closest point for (%v, %v): %v\n", x, y, closest_id)
    number_of_mins := countMin(dists)

    if number_of_mins > 1{
        closest_id = 99
    }
    
    // check now if smallest distance is shared between 2 points
    return closest_id
}
```

```go
// iterate over field to get closest point
rows, cols := field.Dims()
for x:=0; x < cols; x++{
    for y:=0; y < rows; y++{
        closest := FindClosestPoint(x, y)
        field.Set(int(y), int(x), closest)
    }
} 

b := field.T()
aT := mat.DenseCopyOf(b)
printField(aT)
```

```go
// only keep the ones that do not border the array
var first_row []float64 = aT.RawRowView(0)
var last_row []float64 = aT.RawRowView(cols-1)
first_col := field.RawRowView(0)
last_col := field.RawRowView(rows-1)
```

```go
var v []float64
v = append(first_row, last_col...)
v = append(v, last_row...)
v = append(v, first_col...)
```

```go
func contains(s []float64, e float64) bool {
    for _, a := range s {
        if a == e {
            return true
        }
    }
    return false
}
```

```go
var unique []float64
for cur_id, _ := range mapping{
    if !contains(v, cur_id){
        unique = append(unique, cur_id)
    }
}
```

```go
unique
```

```go
// count area
func CountArea(val float64) int{
    cnt := 0
    rows, cols := field.Dims()
    for x:=0; x < cols; x++{
        for y:=0; y < rows; y++{
            if field.At(int(y), int(x)) == val{
                cnt++
            }
        }
    } 
    return cnt
}


var areas []float64
for _, u := range unique{
    areas = append(areas, float64(CountArea(u)))
}

max_field := floats.Max(areas)

fmt.Printf("Max area: %v\n", max_field)
```

### Part (2)

```go
field := mat.NewDense(int(max_y) + 1, int(max_x) + 1, nil)

// init field

for i, coord := range mapping{
    field.Set(int(coord[1]), int(coord[0]), i)
}
```

```go
func CheckIfDistancesUnderThreshPoint(x int, y int, thresh float64) bool{
    dists := []float64{}
    
    for _, coord := range mapping{
        dist := ManhattanDistance(coord, []float64{float64(y), float64(x)})
        dists = append(dists, float64(dist))
    }
    
    sum_dists := floats.Sum(dists)
    if sum_dists < thresh{
        return true
    }

    return false
}
```

```go
// iterate over field to get closest point
rows, cols := field.Dims()
thresh := float64(10000)
cnt := 0
for x:=0; x < cols; x++{
    for y:=0; y < rows; y++{
        
        if CheckIfDistancesUnderThreshPoint(x, y, thresh){
            cnt++
        }
    }
} 

fmt.Printf("Number of points with sum of distances < %v: %v\n", thresh, cnt)
```
